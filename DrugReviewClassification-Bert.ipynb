```python
import pandas as pd
import io

dfTrain = pd.read_csv('C:\\Users\\91988\\Downloads\\Train.csv', index_col=False)
dfTest = pd.read_csv('C:\\Users\\91988\\Downloads\\Test.csv', index_col=False)
```

```python
train_df = dfTrain[['text','labels']]
eval_df = dfTest[['text','labels']]
```

```python
from simpletransformers.classification import ClassificationModel

model = ClassificationModel('bert', 'bert-base-cased', use_cuda = False, num_labels=884, args={'reprocess_input_data': True, 'overwrite_output_dir': True}) 

# Create a TransformerModel
# model = ClassificationModel('roberta', 'roberta-base', use_cuda = False)
```

```python
# Train the model

model.train_model(train_df)
```

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
